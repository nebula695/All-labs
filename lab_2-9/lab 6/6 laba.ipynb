```python
class UserAccount:
  def __init__(self, username, email, password):
    self.username = username
    self.email = email
    self.__password = password

  def set_password(self, new_password):
    self.__password = new_password


  def check_password(self, password):
    if password == self.__password:
      return True
    else:
      return False

user = UserAccount('dolbaeb','kokojambo',1234)

user.set_password(76554746543)
print(user.check_password(1241294120924))



```

```python
class Vehicle:
  def __init__(self, make, model):
    self.make = make
    self.model = model

  def get_info(self):
    return f'марка: {self.make}, модель: {self.model}'


class Car(Vehicle):
  def __init__(self, make, model, type_toplivo):
    super().__init__(make, model)
    self.type_toplivo = type_toplivo
  def get_info(self):
    info = super().get_info()
    return f"{info}, ТОПЛИВО: {self.type_toplivo}"

car = Vehicle('r', 't')
print(car.get_info())
ca = Car('r','t', 'fdsjhbvjhb')
print(ca.get_info())

```
